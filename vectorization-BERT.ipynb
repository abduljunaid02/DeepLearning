```python
sentences =[
    ["Hello", "word"],
    ["My", "Name", "is", "Junaid"],
    ["I", "am", "learning",  "word", "vectorization"]
]
sentences
```

```python
from gensim.models import Word2Vec

# Train the Word2Vec model
model = Word2Vec(sentences, min_count=1)

```

```python
# Print the word vector for a word (e.g., 'machine')
word_vector = model.wv['word']
print("Word vector for 'machine':", word_vector)
print(len(word_vector))
```

```python
import numpy as np
```

```python
def get_sentence_vector(sentences, model):
    word_vectors = [model.wv[word] for word in sentences if word in model.wv]
    print(len(word_vectors))
    if len(word_vectors) == 0:
        return np.zeros(model.vector_size)
    return np.mean(word_vectors, axis = 0)
```

```python
sentence = ["Dippemaus", "is", "thrilling"]
sentence_vector = get_sentence_vector(sentence, model)
print("Sentence vector is ", sentence_vector)
print("Length of Sentence vector is ", len(sentence_vector))
```

```python
from transformers import BertModel, BertTokenizer
```

```python
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertModel.from_pretrained("bert-base-uncased")
```

```python
sentence = "The summers is Germany are exciting"

sentToken = tokenizer(sentence, return_tensors="pt")
```

```python
import torch
with torch.no_grad():
    outputs = model(**sentToken)
```

```python
last_hidden_states = outputs.last_hidden_state
len(last_hidden_states[0][0])
```

```python
last_hidden_states.size()
```

```python
sentence_embedding = last_hidden_states.mean(dim = 1).squeeze()
sentence_embedding
```

```python
#CLS EMBEDDING

cls_embedding = last_hidden_states[:,0,:]
print("The shape of CLS embedding is ", cls_embedding.shape)
print("The CLS vector is ",cls_embedding)
```

```python
#Exploring the similarity of the sentences

from transformers import BertModel, BertTokenizer
import torch
from torch.nn.functional import cosine_similarity

sent1 = "The administration of telangana has gone really bad"
sent2 = "The previous government of telangana did a much better job at administration"

tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertModel.from_pretrained("bert-base-uncased")

sent1Token = tokenizer(sent1, return_tensors="pt")
sent2Token = tokenizer(sent2, return_tensors="pt")

with torch.no_grad():
    output1 = model(**sent1Token)
sent1embedding = output1.last_hidden_state[:,0,:]

with torch.no_grad():
    output2 = model(**sent2Token)
sent2embedding = output2.last_hidden_state[:,0,:]

similarityIndex = cosine_similarity(sent1embedding, sent2embedding)
print(f"The sentences are {similarityIndex.item():.4f} similar")
```
